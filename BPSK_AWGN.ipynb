```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#실험 반복 횟수 정의
iter=10**5
#송신 파워 범위 정의
p=np.arange(0,16,1)
lin_p=10**(p/10)
#BER 배열 정의
ber=np.zeros(len(p))
```

```python
for j in range(0,len(p)):
    err=0
    for i in range(0,iter):
        #송신 신호 정의
        x=-1+2*np.random.randint(2,size=1)
        #노이즈 정의
        n=np.random.randn()
        if x==1:
            if np.sqrt(lin_p[j])*x+n<0:
                err=err+1
        else:
            if np.sqrt(lin_p[j])*x+n>0:
                err=err+1
    ber[j]=err/iter
```

```python
print(ber)
```

```python
plt.semilogy(p,ber)
plt.show()
```
